```python
# Fill missing text fields with 'Unknown'
text_columns = df.select_dtypes(include='object').columns
df[text_columns] = df[text_columns].fillna('Unknown')

# Fill numeric missing values with median
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
#Data_Validation
```

```python
df.describe()
```

```python
# Check unique values for categorical columns
# Re-initialize text_columns with current object type columns after renaming
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    print(f"{col}:", df[col].nunique())
```

```python
# Convert date columns safely
if 'created_at' in df.columns:
    df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
#Data_transformation
```

```python
#Label_encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in text_columns:
    df[col] = le.fit_transform(df[col])
```

```python
#Text_Cleaning
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text
```

```python
if 'ticket_description' in df.columns:
    df['ticket_description'] = df['ticket_description'].astype(str).apply(clean_text)
```

```python
#Data_Enrichment
```

```python
#Ticket_Length_Feature
df['ticket_length'] = df['ticket_description'].apply(len)
```

```python
#priotrize_Catezorization
def priority_level(x):
    if x > 300:
        return 'High'
    elif x > 100:
        return 'Medium'
    else:
        return 'Low'

df['priority_level'] = df['ticket_length'].apply(priority_level)
```

```python
df['priority_level'] = le.fit_transform(df['priority_level'])
```
